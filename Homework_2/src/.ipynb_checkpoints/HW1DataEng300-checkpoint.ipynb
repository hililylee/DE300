```python
# Import the necessary library
import sqlite3

# Connect to the database
conn = sqlite3.connect('HeartDisease.db')
c = conn.cursor()

# Calculate the mode for each categorical feature
categorical_columns = ["painloc","painexer","relrest", "cp", "sex", "htn","smoke",
                  "fbs", "dm", "famhist", "restecg", "dig", "prop", "nitr",
                  "pro","diuretic", "proto", "exang","xhypo", "slope", "ca",
                  "exerckm","restwm","exerwm","thal","thalsev","thalpul","earlobe"]

update_queries = []
for col in categorical_columns:
    # Get the mode value
    mode_value = HeartDisease[col].mode().iloc[0]
    # Construct the UPDATE query
    update_queries.append(f"UPDATE HeartDisease SET {col} = '{mode_value}' WHERE {col} IS NULL")

# Execute all UPDATE queries for categorical columns
for query in update_queries:
    c.execute(query)

# Update pncaden column with the sum of painloc, painexer, and relrest
c.execute("UPDATE HeartDisease SET pncaden = painloc + painexer + relrest")

# Replace missing values in restckm column with zeroes
c.execute("UPDATE HeartDisease SET restckm = 0 WHERE restckm IS NULL")

# Commit the changes
conn.commit()


# Close the connection
conn.close()
```

Next, I identified which features were categorical by refering to the dataset feature descriptions provided on the homework. I also noticed that one of the columns was empty but the datasheet said that it should be the sum of columns 5,6, and 7, so I went ahead and made the respective changes. In terms of the missing values for categorical features, I chose to impute the missing values with the mode of each column. The mode represents the most frequent value within a category. By imputing missing values with the mode, I aimed to replace them with the value that most likely belongs in that category based on the existing data. Categorical data represents distinct categories so finding the mean or median would not be the best option.

```python
#IDENTIFY OUTLIERS

# Start with data cleaning
for column in HeartDisease.columns:
    HeartDisease[column] = pd.to_numeric(HeartDisease[column], errors='coerce')

# Calculate means and standard deviations (numeric values)
mean = HeartDisease.mean(skipna=True)
std = HeartDisease.std(skipna=True)

# Calculate z-scores using numeric values
z_scores = (HeartDisease - mean) / std

# Set a threshold for Z-score
threshold = 3


# Find outliers using Z-score method
outliers_z = HeartDisease[(z_scores > threshold).any(axis=1)]

# Calculate IQR
Q1 = HeartDisease.quantile(0.25)
Q3 = HeartDisease.quantile(0.75)
IQR = HeartDisease - Q1

# Find outliers using IQR method
outliers_iqr = HeartDisease[((HeartDisease < (Q1 - 1.5 * IQR)) | (HeartDisease > (Q3 + 1.5 * IQR))).any(axis=1)]

print("Outliers detected using Z-score method:")
print(outliers_z)

print("\nOutliers detected using IQR method:")
print(outliers_iqr)
```

For the numerical features, I used 2 methods, Z-score and IQR (Interquartile Range),to identify outliers in the data in order to determine whether I should impute the missing values with the mean or median. Based on these methods, I noticed that the data contains a significant number of outliers, so the mean can be heavily influenced by these extreme values, leading to biased imputations. In this case, the median is a more robust measure of central tendency because it is less affected by outliers.

```python
#IMPUTE MISSING INTEGERS WITH MEAN

# Connect to the database
conn = sqlite3.connect('HeartDisease.db')
c = conn.cursor()

# Numerical columns
numerical_columns = ["age", "trestbps","chol","cigs","years",
            "ekgmo", "ekgyr", "ekgday","thaldur","thaltime","met","thalach","thalrest","tpeakbps",
            "tpeakbpd","dummy","trestbpd","oldpeak", "rldv5","rldv5e","restef", "exeref", "restckm",
                     "cmo","cday","cyr"]

# Iterate over each numerical column
for col in numerical_columns:
    # Calculate the mean value using SQL
    c.execute("SELECT AVG({}) FROM HeartDisease".format(col))
    mean_value = c.fetchone()[0]
    # Replace missing values with the mean value
    c.execute("UPDATE HeartDisease SET {} = COALESCE({}, {}) WHERE {} IS NULL".format(col, col, mean_value, col))

# Commit the changes
conn.commit()

# Close the connection
conn.close()
```

For the numerical features, I chose to impute the missing integers with the mean values for each column. I wanted to ensure that the overall distribution and central tendency of the data remain relatively unchanged.

```python
#REMOVE SIGNIFICANT OUTLIERS

# Connect to the database

conn = sqlite3.connect('HeartDisease.db')
c = conn.cursor()
# Start with data cleaning
for column in HeartDisease.columns:
    HeartDisease[column] = pd.to_numeric(HeartDisease[column], errors='coerce')

# Calculate means and standard deviations (numeric values)
mean = HeartDisease.mean(skipna=True)
std = HeartDisease.std(skipna=True)

# Calculate z-scores using numeric values
z_scores = (HeartDisease - mean) / std

# Set a threshold for Z-score
threshold = 6


# Find outliers using Z-score method
outliers_z = HeartDisease[(z_scores > threshold).any(axis=1)]


# Remove outliers from the DataFrame
HeartDisease = HeartDisease[~HeartDisease.index.isin(outliers_z.index)]

# Commit the changes
conn.commit()

# Close the connection
conn.close()
```

I decided to remove outliers based on the Z-score method with a threshold of 6. I wanted to be conservative in my outlier detection because this approach minimizes the risk of discarding valid data points that might be crucial for accurate analysis, especially if the underlying distribution is not perfectly normal.

```python
#FEATURE TRANSFORMATION

# Assuming HeartDisease is your DataFrame and numerical_columns is a list of numerical column names
for column in numerical_columns:
    data = HeartDisease[column].dropna()  # Remove NaN values if any
    
    if data.min() <= 0:
        # Handle zero or negative values by adding a small constant
        min_nonzero = data[data > 0].min() if (data > 0).any() else 1  # Get the smallest nonzero value
        data = data + min_nonzero / 100  # Add a small fraction of the smallest nonzero value
    
    kurt = stats.kurtosis(data)
    skew = stats.skew(data)
    
    if kurt > 5 or skew > 1:
        # Feature is not normally distributed
        # Perform transformations
        
        transformed_data = np.log1p(data)
        
        # Filter out invalid values after transformation
        transformed_data = transformed_data[~np.isnan(transformed_data)]
        
        if len(transformed_data) > 0:
            # Plot histogram and Q-Q plot
            fig, axes = plt.subplots(1, 2, figsize=(12, 5))

            sns.histplot(transformed_data, kde=True, ax=axes[0])
            axes[0].set_title(f'Histogram of {column} (Transformed)')

            stats.probplot(transformed_data, dist="norm", plot=axes[1])
            axes[1].set_title(f'Q-Q plot of {column} against normal (Transformed)')

            plt.show()

            # Update DataFrame with transformed data
            HeartDisease[column] = transformed_data
        else:
            print(f"Warning: Transformed data for column '{column}' is empty after filtering out NaN values.")
        
    else:
        # Feature is normally distributed
        pass  # Do nothing
```

In order to perform feature transformation, I used kurtosis and skewness to identify which features are not normally distributed. If one of the values were large, I plotted a histogram and a Q-Q plot vs normal to check again to see if the feature is normal or not. For features that were not normal, I performed log transformation. 

```python
# FEATURE TRANSFORMATION - ROUND TO 2 DIGITS
conn = sqlite3.connect('HeartDisease.db')
c = conn.cursor()

numerical_columns = ["age", "trestbps","chol","cigs","years",
            "ekgmo", "ekgyr", "ekgday","thaldur","thaltime","met","thalach","thalrest","tpeakbps",
            "tpeakbpd","dummy","trestbpd","oldpeak", "rldv5","rldv5e","restef", "exeref", "restckm",
                     "cmo","cday","cyr"]

# Round numerical columns to two digits after the decimal point
for col in numerical_columns:
    c.execute("UPDATE HeartDisease SET {} = ROUND({}, 2)".format(col, col))

# Commit the changes
conn.commit()

# Fetch and print the rows
c.execute("SELECT * FROM HeartDisease")
rows = c.fetchall()
for row in rows:
    print(row)

# Close the connection
conn.close()
```

```python
#BOXPLOTS FOR NUMERICAL FEATURES

def plot_boxplots(data, numerical_columns):
  # Calculate the number of rows and columns needed for subplots
  num_plots = len(numerical_columns)
  num_rows = (num_plots + 3) // 4  # Round up division
  num_cols = min(num_plots, 4)

  # Create subplots
  fig, axes = plt.subplots(num_rows, num_cols, figsize=(16, 4*num_rows))

  # Flatten axes if necessary
  if num_plots > 1:
      axes = axes.flatten()

  # Create boxplot for each non-empty column
  valid_plot_indices = []
  for i, column in enumerate(numerical_columns):
      if HeartDisease[column].dropna().shape[0] > 0:  # Check if column is not empty
          sns.boxplot(data=HeartDisease[column], ax=axes[i])
          axes[i].set_title(f'Boxplot of {column}')
          axes[i].set_xlabel(column)
          valid_plot_indices.append(i)

  # Hide unused subplots
  for j in range(len(valid_plot_indices), num_rows*num_cols):
      fig.delaxes(axes[j])

  plt.tight_layout()
  plt.show()
    
# Application

numerical_columns_1 = ["age", "trestbps","chol","cigs"]
numerical_columns_2 = ["years","ekgmo", "ekgyr", "ekgday"]
numerical_columns_3 = ["thaldur","thaltime","met","thalach"]
numerical_columns_4 = ["thalrest","tpeakbps","tpeakbpd","dummy"]
numerical_columns_5 = ["trestbpd","oldpeak", "rldv5","rldv5e"]
numerical_columns_6 = ["restef", "exeref", "cmo","cday","cyr"]

plot_boxplots(HeartDisease, numerical_columns_1)
plot_boxplots(HeartDisease, numerical_columns_2)
plot_boxplots(HeartDisease, numerical_columns_3)
plot_boxplots(HeartDisease, numerical_columns_4)
plot_boxplots(HeartDisease, numerical_columns_5)
plot_boxplots(HeartDisease, numerical_columns_6)
```

```python
#BOX PLOTS FOR CATEGORICAL FEATURES

def plot_barcharts(data, categorical_columns, filter_column='target', filter_value=1):
    num_cols = len(categorical_columns)
    num_rows = (num_cols - 1) // 4 + 1

    # Create a figure with subplots
    fig, axes = plt.subplots(nrows=num_rows, ncols=4, figsize=(14, num_rows * 3))

    # Flatten the axes array for easy iteration
    axes_flat = axes.flatten()

    # Find the maximum count across all columns
    max_count = 0
    for col in categorical_columns:
        filtered_data = data[data[filter_column] == filter_value]
        count = filtered_data[col].value_counts().max()
        max_count = max(max_count, count)

    # Loop through each column and create a bar chart on its corresponding subplot
    for i, col in enumerate(categorical_columns):
        filtered_data = data[data[filter_column] == filter_value]
        sns.countplot(data=filtered_data, x=col, ax=axes_flat[i])

        # Set title for each subplot
        axes_flat[i].set_title(col)

        # Set common y-axis limit for all subplots
        axes_flat[i].set_ylim(0, max_count + 2)  # +2 for a bit of padding

    # Adjust layout
    plt.tight_layout()

    # Display the plot
    plt.show()

plot_barcharts(HeartDisease, categorical_columns)
```

These bar charts show the count for each of these categorical features among those who had > 50% diameter narrowing diagnosis of heart disease. Based on the bar charts, most of the people who had this diagnosis experienced chestpain, experienced relief after rest, had asymptomatic chest pain, and were male. Additionally, a majority of those with diagnosis did not have digitalis used furing exercise (dig), nor calcium channel blocker and diuretic used during exercise (pro and diuretic). The most common slope of the peak exercise ST segment was flat slope (slope=2).
