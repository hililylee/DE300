```python
import requests
from bs4 import BeautifulSoup

# Extract content from source 1

# Send a GET request to the webpage
url1 = 'https://www.abs.gov.au/statistics/health/health-conditions-and-risks/smoking-and-vaping/latest-release'
response1 = requests.get(url1)

# Parse the HTML content
soup1 = BeautifulSoup(response1.text, 'html.parser')

# Find the table based on its class name
table1 = [table for table in soup1.find_all('table', class_='responsive-enabled') if table is not None][1]

# Create a dictionary of smoking rates based on age

smoking_rates = []

# start with ages 15-74

for i in range(1, 8):
  smoke_data_cells = table1.find_all('tr')[i].find_all('td')[9]
  smoke_rate = smoke_data_cells.get_text(strip=True)

  # Convert the smoke rate to a float and round it to the nearest integer
  smoke_rate = float(smoke_rate)

  #Parse the age range and find the age values in between the age range
  age_range_data_cells = table1.find_all('th', {'scope': 'row', 'class': 'row-header'})[i-1]
  age_range = age_range_data_cells.get_text(strip=True)
  age_values = age_range.replace('â€“', '-').split('-')
  age_start = int(age_values[0])
  age_end = int(age_values[1])
  age_sequence = range(age_start, age_end + 1)

  #Return the data
  data_tuple = tuple(age_sequence) + (smoke_rate,)

  # Create a dictionary to map each age to its smoke rate
  age_to_smoke_rate = {age: smoke_rate for age in age_sequence}

  # Append data to the smoking_rates list
  smoking_rates.append((age_to_smoke_rate,))
```

```python
# parse out the smoke rate for those 75 and above

senior_smoke_data_cells = table1.find_all('tr')[8].find_all('td')[9]
senior_smoke_rate =senior_smoke_data_cells.get_text(strip=True)
senior_smoke_rate = float(senior_smoke_rate)
senior_age_range_data_cells = table1.find_all('th', {'scope': 'row', 'class': 'row-header'})[7]
senior_age_range = senior_age_range_data_cells.get_text(strip=True)
senior_age = int(senior_age_range.split()[0])
```

```python
# Update database with smoking rates from source 1

# Connect to the SQLite database
conn = sqlite3.connect('HeartDisease2.db')
c = conn.cursor()

# Iterate over the data and update the database
for age_to_smoke_rate_tuple in smoking_rates:
    age_to_smoke_rate_dict = age_to_smoke_rate_tuple[0]  # Extract the dictionary from the tuple
    for age, smoking_rate in age_to_smoke_rate_dict.items():
        c.execute("UPDATE HeartDisease2 SET smokesource1 = ? WHERE age = ?", (smoking_rate, age))
        c.execute("UPDATE HeartDisease2 SET smokesource1 = ? WHERE age >= ?", (senior_smoke_rate, senior_age))


# Commit the changes to make them permanent
conn.commit()

# Close the connection
conn.close()
```

```python
# Extract content from source 2

# Send a GET request to the webpage
url2 = 'https://www.cdc.gov/tobacco/data_statistics/fact_sheets/adult_data/cig_smoking/index.htm'
response2 = requests.get(url2)

# Parse the HTML content
soup2 = BeautifulSoup(response2.text, 'html.parser')

# Find the list with class block-list
block_list = soup2.find_all('ul', class_='block-list')[2]

# Extract the text content of the first list item
desired_line1 = block_list.find_all('li')[0].text.strip()
desired_line2 = block_list.find_all('li')[1].text.strip()
desired_line3 = block_list.find_all('li')[2].text.strip()
desired_line4 = block_list.find_all('li')[3].text.strip()

def extract_numbers(text):
    # Extracting numbers based on specific patterns
    numbers = []
    for word in text.split():
        if word.isdigit():  # Check if the word is a digit
            numbers.append(word)
        elif 'â€“' in word:   # Check for en dash representing age range
            ages = word.split('â€“')
            numbers.extend(ages)
        elif '%' in word:   # Check for percentage
            numbers.append(word.strip('()%'))
    return numbers
```

```python
# Update smoke rate for FEMALES using source 2

# Connect to the SQLite database
conn = sqlite3.connect('HeartDisease2.db')
c = conn.cursor()

#Age group 1
group1_data = extract_numbers(desired_line1)
c.execute("UPDATE HeartDisease2 SET smokesource2 = ? WHERE sex = 0 AND age >=? AND age <=?", (group1_data[4],group1_data[2], group1_data[3]))

#Age group 2
group2_data = extract_numbers(desired_line2)
c.execute("UPDATE HeartDisease2 SET smokesource2 = ? WHERE sex = 0 AND age >=? AND age <=?", (group2_data[4],group2_data[2], group2_data[3]))

#Age group 3
group3_data = extract_numbers(desired_line3)
c.execute("UPDATE HeartDisease2 SET smokesource2 = ? WHERE sex = 0 AND age >=? AND age <=?", (group3_data[4],group3_data[2], group3_data[3]))

#Age group 4
group4_data = extract_numbers(desired_line4)
c.execute("UPDATE HeartDisease2 SET smokesource2 = ? WHERE sex = 0 AND age >=?", (group4_data[3],group4_data[2]))

# Commit the changes to make them permanent
conn.commit()

# Close the connection
conn.close()
```

```python
# Transform the smokesource1 and smokesource2 columns

# Both graphs showed an exponential distribution so I am applying a logarithmic 
#transformation to make the distribution closer to a normal distribution

# Connect to the SQLite database
conn = sqlite3.connect('HeartDisease2.db')
c = conn.cursor()

# Define a function for logarithmic scaling
def log_scaling(x):
    return np.log(x + 1)  # Adding 1 to avoid log(0)

# Update smokesource1
c.execute("SELECT smokesource1 FROM HeartDisease2")
rows = c.fetchall()
for row in rows:
    updated_value = log_scaling(row[0])
    c.execute("UPDATE HeartDisease2 SET smokesource1 = ? WHERE smokesource1 = ?", (updated_value, row[0]))

# Update smokesource2
c.execute("SELECT smokesource2 FROM HeartDisease2")
rows = c.fetchall()
for row in rows:
    updated_value = log_scaling(row[0])
    c.execute("UPDATE HeartDisease2 SET smokesource2 = ? WHERE smokesource2 = ?", (updated_value, row[0]))


# Commit the changes to make them permanent
conn.commit()

# Close the connection
conn.close()
```

```python
# The 2nd graph which represents smokesource2 also showed a bimodal distribution so I am applying standard scaling transformation.

# Connect to your database
conn = sqlite3.connect('HeartDisease2.db')
c = conn.cursor()

# Fetch the data from the database
c.execute("SELECT smokesource2 FROM HeartDisease2")
rows = c.fetchall()

# Calculate mean and standard deviation
data = [row[0] for row in rows]
mean = np.mean(data)
std_dev = np.std(data)

# Update the data with standard scaling
for row in rows:
    original_value = row[0]
    updated_value = (original_value - mean) / std_dev
    c.execute("UPDATE HeartDisease2 SET smokesource2 = ? WHERE smokesource2 = ?", (updated_value, original_value))

# Commit the changes
conn.commit()

# Close the connection
conn.close()
```

```python
#Tasks 3-5

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
import numpy as np
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect('HeartDisease2.db')

# Load data from the database
query = "SELECT * FROM HeartDisease2"
data = pd.read_sql_query(query, conn)

# Separate features (X) and target variable (y)
X = data.drop(columns=['target'])  
y = data['target']

# Step 1: Splitting the Data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, stratify=y, random_state=42)

# Step 2: Training Binary Classification Models
models = {
    'Logistic Regression': LogisticRegression(),
    'Random Forest': RandomForestClassifier(),
    'SVM': SVC()
}

for name, model in models.items():
    model.fit(X_train, y_train)

    # Step 3: Assessing Performance with 5-fold cross-validation
    cv_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='accuracy')
    print(f"{name} Cross-Validation Scores: {cv_scores}")
    print(f"{name} Mean Cross-Validation Score: {np.mean(cv_scores)}")

    # Step 4: Evaluating performance metrics
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred)

    print(f"{name} Test Set Metrics:")
    print(f"Accuracy: {accuracy}")
    print(f"Precision: {precision}")
    print(f"Recall: {recall}")
    print(f"F1-score: {f1}")
    print(f"AUC-ROC: {roc_auc}")
    print()

# Step 5: Selecting the Final Model
best_model_name = max(models, key=lambda k: np.mean(cross_val_score(models[k], X_train, y_train, cv=5)))
print(f"Final selected model: {best_model_name}")

# Close the database connection
conn.close()
```

The final selected model was logistic regression. It had the highest mean cross-validation score and is a relatively more interpretable model compared to Random Forests and Support Vector Machines (SVMs). The coefficients of the Logistic Regression model can be used to understand the influence of each feature on the target variable, making it more interpretable. Random Forests and SVMs, on the other hand, are considered black-box models, so it is difficult to understand the underlying decision-making process and the contribution of each feature to the final prediction. Based on the metrics and the requirement for interpretability, the Logistic Regression model would be the most appropriate selection as the final model for this heart disease prediction task.
