```python
def get_info(pd_frame):
    res = pd_frame.describe().transpose()
    res['missing_percent'] = [(pd_frame.shape[0] - res.loc[ind, 'count'])/pd_frame.shape[0] for ind in res.index]
    res['unique_val'] = [len(pd_frame.loc[:,ind].value_counts()) for ind in res.index]
    res['var_type'] = ['Continuous' if res.loc[ind, 'unique_val'] >=100 else 'Categorical' for ind in res.index]
    return res
```

```python
jan_info = get_info(jan)
jan_info
```

## # of Flights vs State

```python
# Count the occurrences of each state in the 'origin' column
top_states = jan['OriginState'].value_counts().head(5)

# Plot a barchart
plt.figure(figsize=(10, 6))  # Adjust the width and height as needed
top_states.plot(kind='bar')
plt.xlabel('State')
plt.ylabel('Number of Flights')
plt.title('Top 5 States People Fly Out From')
plt.xticks(rotation=0)  
plt.tight_layout()
plt.show()
```

By creating a bar chart of total flight counts versus origin state, I could visualize which states had the highest number of outgoing flights. The flight count could be attributed to either a large number of people flying out regularly or a smaller number of people flying frequently. Further data analysis would need to be conducted to determine this attribute. 

Based on the bar chart showcasing the top 5 states, I noticed that most of these states are states with high population densities. In fact, according to the recent population statistics from US News and World Report, the top 5 states with the highest number of flights shown in this bar chart are among the top 8 most populated states in the United States. Moreover, these states are known to have booming business sectors, so there may be a significant number of business travelers flying out for work trips.

## Distance Flown vs Day of the Week

```python
# creating a categorical column for weekday
jan['weekday'] = (
    jan['DayOfWeek']
    .replace({1:'Mon',2:'Tues',3:'Wed',4:'Thurs',5:'Fri',6:'Sat',7:'Sun'})
    .astype('category')
)
```

```python
average_distance_per_day = jan.groupby('DayOfWeek')['Distance'].mean()
```

```python
print (average_distance_per_day)
```

```python
# Calculate the width of each bar
bar_width = 0.5  # Adjust as needed
bar_positions = np.arange(len(average_distance_per_day.index))

# Plot the bars with adjusted x-coordinates
plt.bar(bar_positions, average_distance_per_day, width=bar_width)

# Set x-axis tick positions and labels
plt.xticks(bar_positions, ['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun'])

plt.xlabel('Day of the Week')
plt.ylabel('Average Distance Flown')
plt.title('Average Distance Flown per Day of the Week')
plt.show()
```

```python
# Create a figure and axes
fig, axs = plt.subplots(1, 7, figsize=(15, 5))  # Adjust the width and height as needed

# Create a box plot for each day of the week
for i, day in enumerate(['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun']):
    sns.boxplot(data=jan[jan['weekday'] == day], y='Distance', ax=axs[i], width=0.5)  # Adjust width as needed
    axs[i].set_title(f'Distance on {day}')

# Show the plot
plt.tight_layout()
plt.show()
```

In order to examine the relationship between flight distance and the day of the week, I created a bar chart and boxplots of distances for each day. When creating the bar chart, I compared the average distance flown versus the day of the week. Since there may be a different number of flights on each day of the week, comparing the average distance would be more accurate than using the total distance. Based on the bar chart, I did not notice a significant difference between the average distance flown for each day of the week. However, it seems that on Saturday, the average distance flown is slightly greater than the rest of the days. This makes sense because Saturday is the start of the weekend when most people have time off to travel. Meanwhile, Wednesday, which is in the middle of the work week, seems to have the lowest average distance. To increase revenue, airline companies may consider offering incentives, such as double points or discounts, on Wednesdays. From the boxplots of the distances for each day, I noticed a significant number of upper outliers, especially closer towards the maximum of each boxplot. For the most part, the boxes themselves seem to show a normal distribution, but there are many flights with much longer distances, which may be international flights. While most flights are likely domestic, as shown by the normal distribution, the international flights are usually much longer distances and may be the extreme values.
